#画像の読み込み、顔検出処理

```
img = cv2.imread("/content/drive/My Drive/OpenCV/face2.jpg")

# 画像のRGB形式への変換
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
# 解析ライブラリの使用
cascade = cv2.CascadeClassifier("/content/drive/My Drive/OpenCV/haarcascade_frontalface_alt.xml")
# 検出処理
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
face = cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=3, minSize=(30, 30))
# 検出領域のデザイン処理
for (x, y, w, h) in face:
  cv2.rectangle(img, (x, y), (x + w, y + h), (200,0,0), 3)
# 標準軸の非表示処理
plt.grid(False)
plt.tick_params(labelbottom=False, labelleft=False, labelright=False, labeltop=False)
plt.imshow(img)
```

```
#画像の切り抜き
face = img[y: y+h, x: x+w]

# 標準軸の非表示処理
plt.grid(False)
plt.tick_params(labelbottom=False, labelleft=False, labelright=False, labeltop=False)
plt.imshow(face)
```

```
#画像の切り抜き ※intは小数点以下切り捨て
cheek_R = img[int(y+h*11/20): int(y+h*13/20), int(x+w*2/10): int(x+w*3/10)]

# 標準軸の非表示処理
plt.grid(False)
plt.tick_params(labelbottom=False, labelleft=False, labelright=False, labeltop=False)
plt.imshow(cheek_R)
```
